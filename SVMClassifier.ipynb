```python
X = df.drop("loan_risk", axis = 1)
y = df["loan_risk"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, random_state= 15, test_size= 0.2
)
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC(kernel= 'linear')
```

```python
svc.fit(X_train, y_train)
```

```python
y_pred = svc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
print("score: ", accuracy_score(y_pred, y_test))
print("matrix: ", confusion_matrix(y_pred, y_test))
print("classification: ", classification_report(y_pred, y_test))
```

```python
svc = SVC(kernel= 'linear', C= 10, class_weight= {0:1, 1:5})
svc.fit(X_train, y_train)
y_pred = svc.predict(X_test)
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
print("score: ", accuracy_score(y_test, y_pred))
print("matrix: ", confusion_matrix(y_test, y_pred))
print("classification: ", classification_report(y_test, y_pred))
```

```python
svc = SVC(kernel= 'rbf', C= 10, class_weight= {0:1, 1:5})
svc.fit(X_train, y_train)
y_pred = svc.predict(X_test)
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
print("score: ", accuracy_score(y_test, y_pred))
print("matrix: ", confusion_matrix(y_test, y_pred))
print("classification: ", classification_report(y_test, y_pred))
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

x_min, x_max = X_train.iloc[:, 0].min() - 1, X_train.iloc[:, 0].max() + 1
y_min, y_max = X_train.iloc[:, 1].min() - 1, X_train.iloc[:, 1].max() + 1

xx, yy = np.meshgrid(
    np.linspace(x_min, x_max, 300),
    np.linspace(y_min, y_max, 300)
)

grid = np.c_[xx.ravel(), yy.ravel()]
Z = svc.decision_function(grid).reshape(xx.shape)

plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, levels=20, alpha=0.3)
plt.contour(xx, yy, Z, levels=[-1, 0, 1], linestyles=['--', '-', '--'])

sns.scatterplot(
    x=X_train.iloc[:, 0],
    y=X_train.iloc[:, 1],
    hue=y_train
)

plt.xlabel(X_train.columns[0])
plt.ylabel(X_train.columns[1])
plt.title("SVM Margin and Decision Boundary")
plt.show()
```

```python
c_values = [100, 10, 1.0, 0.1, 0.01]
class_weight = [{0:w, 1:y} for w in [1,10,50,100] for y in [1,10,50,100]]
kernel = ["rbf", "linear", "poly"]
gamma = ["scale", "auto"]
```

```python
params = dict(C = c_values, class_weight = class_weight, kernel = kernel, gamma = gamma)
```

```python
from sklearn.model_selection import GridSearchCV, StratifiedKFold
```

```python
grid = GridSearchCV(estimator= SVC(), param_grid= params, cv = 5)
```

```python
grid.fit(X_train, y_train)
```

```python
df.shape
```

```python
y_pred_grid = grid.predict(X_test)
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
print("score: ", accuracy_score(y_test, y_pred_grid))
print("matrix: ", confusion_matrix(y_test, y_pred_grid))
print("classification: ", classification_report(y_test, y_pred_grid))
```
