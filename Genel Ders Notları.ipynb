## Support Vector Machines (SVM)

Tanım: SVM, hem sınıflandırma hem de regresyon problemlerinde kullanılan küçlü bir ML algoritmasıdır. 

İki veri kümesi düşün ayrı ayrı, bu veri kümesinin tam ortasından çekilen çizgi 2'ye ayırır bu kümeyi.
Çizginin sağında ve solundaki en yakın veri kümesinden çekilen çizgi(destek vektörü) ile ortadaki best fit line aralığına margin aralığı denir.

Margin aralığının max olduğu aralığa ulaşmak gerekir. Dolayısıyla margini max etmek ve w'yı minimize etmektir.



#### Vektörel Gösterim

Doğrusal modellerde gösterim y = A0+A1X idi. Ancak SVM'de w^T * X + b = 0 <p>

w^T --> Ağırlıklar vektörünün transpozesi. Matrix çarpımı yapıyor yani, çarpım yapabilmek için transpoze etmesi gerekiyor. <p>

X --> girdi özelliklerinin vektörü <p>

b --> sabit terim veya bias <p>

### Hyperparametreler

#### C Values 

Küçük C (0.01, 0.1) <p>

Daha fazla hataya izin verir, overfitting düşük underfitting yüksek olabilir. Daha smooth karar sınırı olur. Yani ana mantık: her nokta doğru sınıflandırılsın mı yoksa biraz hata kabul edilsin mi. Soft veya hard margin konusu burada geçerlidir. 

##### Soft Margin

Destek vektörlerin iç kısımı, yani margin aralığındaki boşluğa herhangi bir veri girmesini kabul eden durumdur. Gerçek dünyaya daha yakın olan margin tipi budur.

##### Hard Margin

Margin arasına herhangi bir noktanın girişinin kabul edilmesi anlamına gelir.

SVC: Support Vector Classifier --> Margin boşluğunu maksimize etmeye çalışır. <p>
SVR: Support Vector Regression --> Lineer regresyonda mantık, best fit line'a yakınlık olmasıydı, burada da best fit lineın iki tarafından da bir çizgi çizerek tüp oluşturur. Bu tüpün içinde kalan alan baz alınır.



#### Kernel Trikleri



Yukarıdaki gibi bir dataset düşünelim, veriler iç içe birbiriyle iç içe girmiş yani. Kernel trik, veri noktalarının ya üssünü alarak ya da matematik fonksiyonu kullanarak boyutları yükseltir ve sağdaki hale getirir. Böylece yüksek başarı ile iki veri birbirinden ayrılmış olur.

**linear <p>
-- Target değişkende linear bir ilişki var ise kullanılır, büyük veri setlerinde hızlıdır.

** rbf <p>
-- Non-linear ilişkileri yakalar.

#### Gamma (rbf, poly ve sigmoid için)

Tanım: tek bir noktanın etki alanını belirler.

***Küçük gamma (0.001) <p>
-- Daha geniş etki alanı olur ve boundary smoothtur. Verilerin birbirine karşı etki alanını belirtir. Örneğin gamma küçük ise en uzaktaki iki veri bile birbirini etkiler.

***Büyük gamma (1,10) <p>
-- Noktalar daha lokal etkilidir, boundry kıvrımlı olur ve verilerin tam uyum sağlamasına force eder. Dolayısıyla overfitting riski yüksektir.

## K-Nearest Neighbors (KNN) Algoritması

Tanım: K-En Yakın Komşu (KNN) algoritması, makine öğrenmesinde hem sınıflandırma hem de regresyon problemleri için kullanılan basit ama etkili bir algoritmadır.

### KNN ile Sınıflandırma

#### Temel Fikir

KNN algoritmasının sınıflandırma amacıyla kullanımı aşağıdaki şekilde özetlenir:

1- Bir k değeri (komşu sayısı) belirlenir. <p>
2- Yeni gelen test verisi için eğitim veri setindeki tüm örneklerle mesafe hesaplanır. <p>
3- Mesafesi en yakın k örnek seçilir. <p>
4- Bu k komuşunun sınıf etkiketleri sayılarak en sık görülen sınıf tahmin edilir <p>



### Mesafe Ölçüm Seçenekleri



1- Manhattan Distance (L1): X ve Y ile oynanır. Ölçülmek istenilen nokta ile eşlenik noktanın x ve y farkları alınıp toplanır. <p>
2- Euclidean Distance (L2): Öklid uzunluğu alınarak direkt ilgili noktaya gidilir. Daha kısa sürüyor.

### Uygulamalı Örnek 



Sonuç: En yakın 3 komuşudan 2 tanesi "Evet" 1 tanesi "Hayır" olduğuna göre, tahmin: "EVET"

### KNN ile Regresyon

#### Temel Fikir

Sınıflandırma yönteminde, en yakın komşular bulunup en çok komşunun hangi sınıfta olduğu bulunurken, regresyonda komşuların ortalaması (outlier varsa medyanı) alınarak tahmin değeri bulunur. 

### KNN Optimizasyonu: KD-Tree ve Ball Tree

KNN algoritmasında tüm veriler için mesafe hesaplandığından bu işlem büyük veri setlerinde maliyetli olabilir.

Örnek: 0-100 arası bir sayı tutalım. Tuttuğumuz sayıya ulaşmak için, tek tek sayıları sayarsak 100. sayıda bulabiliriz en kötü ihtimal, ya da bölerek atıyorum 50 diyip yaklaşarak daha kısa sürede bulabiliriz.

#### KD-Tree

Verileri k boyutlu bir uzayda bölerek ikili bir ağaç yapısında saklar.

Nasıl Çalışır?

1.İlk olarak tüm veriler bir düzlemde sıralanır.<p>
2.İlk bölme, ilk özelliğin medyanına göre yapılır.<p>
3.İkinci düzlem, ikinci özelliğin medyanına göre yapılır.<p>
4.Bu işlem sırayla tüm eksenler arasında geçiş yaparak tekrarlanır.<p>
5.Ağaç oluşturulduktan sonra, sorgu noktası sadece ilgili bölgelerdeki alt kümelerle karşılaştırılır.<p>

#### Ball-Tree

Verileri küresel gruplar halinde hiyerarşik olarak organize eder.

Nasıl Çalışır?

1.Benzer noktalar kümelere ayırılır.<p>
2.Her küme kendi merkez noktasına ve yarıçapına sahiptir.<p>
3.Yeni bir sorgu geldiğinde, sadece yakın kümeler kontrol edilir.<p>
4.Uzakta kalan kümeler tamamen atılabilir<p>

Avantajı: Özellikle yüksek boyutlu veri kümelerinde, Ball Tree yapılması KD Tree'ye göre daha avantajlı olabilir.

### KNN için Uygunluk

KNN algoritması, aşağıdaki durumlarda iyi çalışır:

-- Veri boyutları çok yüksek değilse <p>
-- Özellikler benzer ölçekteyse (scaling ve feature engineering yapıldıysa)
-- Sınıflar arasındaki ayrım netse

### Temel Parametreleri

#### Komşu Sayısı (k)

n_neighbors olarak geçer. En kritik parametredir çünkü K sayısı az olduğunda model karmaşıklaşır ve overfit eder. Büyük k değeri ise pürüzsüz karar sınıfları oluşturarak underfit eder.

#### Mesafe Ölçümü

metric olarak geçer. default olarak minkowski gelir, öklidi uygular.

#### Weights

komşuların etkisini belirler. standart olarak uniform gelir ve her komşu eşit sayıda etki eder. ancak distance yapıldığı durumda daha yakın komşular, dataya daha fazla etki eder.

### Sonuç

KNN basit yapısına rağmen etkili bir algoritma. Yüksek veri setlerinden ziyade daha küçük ölçekli veri setlerinde daha başarılı. Genellikle Öklid'e göre bir yakınlık hesaplandığından kaynaklı scaling ve feature engineering yapılması kritik önem taşımakta.



## Karar Ağaçları (Decision Tree)

Tanım: Karar ağaçları hem sınıflandırma hem de regresyon problemlerinde kullanılabilen, sezgisel ve güçlü bir alogirtmadır. If - else yapısına benzeyen çalışma mantığı vardır, süreç her bir dalın sonunda bir tahmine ulaşana kadar devam eder.

-- En üstteki başlangıç sorusuna root node, yani kök denir. Internal ise nodelar arasındaki bağlantılardır. En aşağıdaki ise leaf node yani yaprak deniyor.  

#### Avantajları

-- Sonuçların nereden alındığını anlamak çok kolaydır. Bu sebeple white - boxs model olarak adlandırılır ve verilerin ölçeklenmesine (standardizasyon) ihtiyaç duymazlar.

#### Dezavantajları

-- En büyük sorunu, eğitim verilerini ezberleyerek overfit edip yeni verilerde başarısız olmasıdır. Bu durumu engellemek adına ağacın derinliğini sınırlandırmak (pruning, budama) gibi yöntemler kullanılır.

#### CART ve ID3

CART: Sadece ikili ağaçların üretilmesine dayanır, evet hayır şeklinde sınıflandırma yapılır. Hem regresyon hem sınıflandırma problemlerinde kullanılabilir.

CART, regresyonda mse değerinin en düşüğünü alarak ikiye ayrılır. 

ID3: CART'tan farklı olarak ID3 ikiden fazla dalı olan düğümler üretebilir. Bu algoritma genellikle kateogrik özelliklerin tüm olası değerleri için ayrı dallar açarak ağacı büyütür. 





### Karar Ağacında Amaç

Karar ağaçlarında amacımız: veri setini öyle bir bölmeliyiz ki, bölme sonucunda oluşan alt kümeler mümkün olduğunca homojen olsun. Bunun için saflık (purity) ölçütleri kullanılır. En yaygın 2 ölçüt <b>(criterion parametresinde kullanılan 2 kriter)</b>:

-- Entropy <p>
-- Gini <p>

### Entropy 

Tanım: Entropy, bilgi teorisinden gelen ve bir veri setindeki düzensizliği ölçen saflık ölçütüdür. Düşük entropy = daha saf küme

Adımlar:

-- İlk olarak amaç en saf sınıfı bulmak. Yani örneğin mülakat dataseti olsun. Örnekte mümkün olduğunca en yüksek mülakattan kabul veya red oranının olduğu örnek seçilmeli. Homojenite ne kadar yüksek ise (yani bütün datalar tek bir sınıfa ait ise) entropi düşer. Tüm datalar tek bir sınıfa ait ise (örneğin tüm datalar mülakatı geçtiyse) entropi 0 olur. <p>

-- Karmaşıklık: Sınıflar düğüm içerisinde ne kadar dengeli dağılmış ise (yani mülakatı geçen kişi sayısı da geçemeyen kişi sayısı da 2 olsun.) entropi o kadar yükselir. 

-- Bilgi Kazancı (Gain Information): Karar ağacı, veriyi bölerken entropiyi en çok azaltan, yani "Bilgi Kazancı" sağlayan özelliği seçer. Bu sebeple Gain Informationı en yüksek olan root node olarak seçilir.




### Gini Impurity

-- Bir karar ağacı düğümündeki verilerin ne kadar "karışık" olduğunu ölçen bir metriktir. Eğer bir düğümdeki tüm örnekşer aynı sınıfa ait ise gini 0 kabul edilir.

-- Entropide olduğu gibi logaritmik işlemler gerekmediğinden daha hızlı çalışır. Default olarak scikit learn kütüphanesinde bu gelir.

### Pruning (Budama)

Ağaç büyüdükçe, overfit olma ihtimali artar. Pruning işlemi ile budama yapılarak ağaç sadeleşir. Ağacın pruning edilmemesi durumunda, eğitim setindeki her ayrıntıyı ve gürültüyü öğrenir ve model çok karışık hale gelir. Budama bu karmaşıklığı azaltır ve modelin yeni verilere daha iyi uyum sağlamasını sağlar. 

Neden overfit eder: 

İlk örnekte value, aldığımız örnekler. İlk aşamada 40, 0, 0 gelmiş. Yani gini 0 olaran true olarak döndürülmüş ve ağaç durmuş.

İkinci aşamada 37 ve 43 olarak diğer gözlemler var. Burada da gözlemleri ayırmaya başlamış. 

En altta 42 gözlemin 38i zaten geçmiş. Yani aslında en aşağıya geldiğinde çoğu şey belli, o aşamadan sonra tekrardan ayırmaya gerek yok. %90'dan fazla zaten karar verilmiş oluyor. Dolayısıyla sonraki aşamalar gereksiz anlamına geliyor.

Yani 120 gözlemde genel bir kural oluşturmak var, bir de her bir örnek için farklı farklı kurallar oluşturmak var. Dolayısıyla çok dallandırırsak overfit edilme ihtimali artar. Pruning, belli bir anlamlı modele ulaştıktan sonra budama yani modeli kesme aşamasıdır.





Temel olarak 2 strateji uygulanır:

#### Ön Budama (Pre-pruning)

Ağacın büyümesini eğitim sırasında belirli kriterlerle durdurur. Örneğin: maksimum derinliği kısaltmak (max_depth) veya bir yaprakta bulunması gereken minimum örnek sayısını belirlemek (min_samples_leaf) birer ön budama işlemidir. 

#### Son Budama (Post-pruning)

Ağaç önce tam derinliğe ulaşana kadar büyür, ardından istatistiksel olarak anlamlı bilgi kazancı sağlamayan (örneğin x2 testi ile p-değeri kontrol edilerek) düğümler silinir veya birleştirilir.

Budama yöntemleri modeli basitleştirerek anlamlı bir model oluşturmaya çalışır.


### Karar Ağaçları ve Regresyon

Sınıflandırma problemlerinde dallanma saflaştırma metrikleri (entropi, gini impurity) ile yapılırken, regresyonda bu metrikler uygun değildir. Bunun yerine;

-- Ortalama Kare Hata (Mean Squared Error - MSE)

-- Varyans Azaltımı (Variance Reduction)

kullanılarak hangi split'in daha iyi olduğu hesaplanır. 

Yani kısaca, varyans veya mse'i en az olanı almaya çalışıyor. 

#### Olası Split Noktalarının Belirlenmesi



 Karar ağacı tüm GPA verileri için ortalamalarını sıralar ve olası split noktaları ortalamalar ile belirlenir.

 


Thresholda göre gruplara ayırılır ve ilk olarak ortalama alınır, daha sonrasında ortalamalar, target değişkenden çıkartılarak varyans hesabı yapılır.

Örneğin sol grup için varyans 11.55 iken sağ grup için 44.19 çıkıyor. Tüm veri seti için varyans (kök varyans) ise 124.83 çıkmaktadır.



En yüksek varyans azaltımı olan eşik seçilerek tahminlemeye başlanır.
