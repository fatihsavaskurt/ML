### Hyperparameter Tuning

#### Logistic Regression Parametreleri

##### 1) Penalty

Tanım: Modelin overfit olmaması için uygulanan ceza türü. <p>
        * l2 --> en yaygını. ridge mantığı direkt, overfit azaltmak için uygulanır. <p>
        * l1 --> lasso. feature selection için kullanılır. lasso ile gereksiz featureslar cezalandırılır. <p>
        * elastic --> ikisinin karışımı <p>

##### 2) C (regularization gücü)

Tanım: En kritik parametrelerden biridir. Modelde overfit veya underfite ne kadar ceza verileceğinin tanımı burasıdır. <p>
    * Overfit var ise --> C küçült <p>
    * Underfit var ise --> C büyült <p>

##### 3) Solver (optimizasyon algoritması)

Tanım: Modelin coef'leri nasıl bulacağını belirtir. <p>
    * liblinear --> küçük dataset, l1 için iyi. <p>
    * lbfgs --> default, genelde kullanılan budur. <p>
    * saga --> büyük dataset + elastic net <p>

##### 4) max_iter

Tanım: Modelin kaç adımda öğreneceğini belirtir. <p>
    İtersyon az ise --> model converge olmaz yani model coefleri yarım öğrenir, dolayısıyla bu durumda arttırılır: 100, 500, 1000 <p>

##### 5) class_weight

Tanım: Azınlığı az cezalandır. <p>

Örnek: <p>
    
    ** Müşterilerin %95'i Hayır desin, %5'i evet desin. Bu durumda başaırya %95 der ancak model çöp. Dolayısıyla class_weight = balanced verilir. Model azınlığı az cezalandırır ve modelin öğrenmesine zorlanır.


##### 6) multi_class

Tanım: Binary değil ise; <p>
    * ovr (one vs rest) --> her sınıf için ayrı model, küçük datasette tercih edilir. <p>
    * multinomial --> tek model, daha sofistike, büyük datasette tercih edilir.





```python
model = LogisticRegression()
```

```python
penalty = ["l1", "l2", "elasticnet"]
c_values = [100, 10, 1, 0.1, 0.01]
solver = ["lbfgs", "liblinear", "netwon-cg", "newton-cholesky", "sag", "saga"]
```

```python
params = dict(penalty=penalty, C = c_values, solver = solver)
```

```python
params
```

```python
from sklearn.model_selection import GridSearchCV, StratifiedKFold
```

```python
cv = StratifiedKFold()
```

```python
grid = GridSearchCV(estimator= model, param_grid= params, cv=cv, scoring = "accuracy", n_jobs= -1)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred = grid.predict(X_test)
```

```python
print("score: ", accuracy_score(y_pred, y_test))
print("matrix: ", confusion_matrix(y_pred, y_test))
print("classification: ", classification_report(y_pred, y_test))
```
