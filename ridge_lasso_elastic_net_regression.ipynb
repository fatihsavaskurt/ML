#### Giriş

-- Ana amaç, best fit lineı bulmak. Eğer direkt olarak noktaların üzerinden gidiyorsa bu çizgi, overfitting problemi ortaya çıkar.

##### Overfitting Problemi

-- Eğer model eğitim verisine çok iyi uyum sağlıyorsa, (yani doğruların tüm noktaların üzerinden geçtiği bir çizgi) ama test verisinde model büyük hatalar yapıyorsa model genelleme yapıyor demektir. Yani o kadar çok eğitiliyor ki X'deki değişkenlere çok fazla tepki vermeye başlıyor demektir.

    -- Eğitim doğruluğu yüksek (R2 score, MSE, MAE, RMSE)
    -- Test doğruluğu düşük (R2 score, MSE, MAE, RMSE)
    -- Bias düşük R2 mae
    -- Varyans yüksek yine yukarıdaki model doğruluğu algoritmaları

##### Ridge Regresyonu (L2 Regularizasyonu) (OVERFIT ENGELLER)

-- Ridge regresyon, doğrusal regresyon modellerine L2 Regularizasyonu ekleyen bir yöntemdir. Bu sayede kat sayılar yani ağırlıklar küçültülerek overfitting engellenmeye çalışılır. 

-- Ridge, lambda parametresiyle bir ceza katsayısı verir. Sola doğru katsayıyı kaydırmak ve global minimayı yukarı doğru çıkarmak. 

-- Modelin katsayıları küçültülmeye zorlanır, bütün değişkenler için küçültülüyor. 

    --Katsayıyı küçülterek underfitting de yapıyor olabilirz. Dolayısıyla katsayıları düşürmeye yarayan lambdayı ayarlamak için cross validation yapmak gerekir. 

-- Yani büyük katsayıları (coefleri) küçültür, küçük katsayıları ise stabilize eder. Böylelikle modelde hem overfittingi hem de multicollinearity(featureların birbirine benzemesi) olayını dengeye çeker.

##### Lasso Regresyonu (L1 Regularizasyonu) (FEATURE SELECTION)

-- 50 kolonlu bir dataset olduğunu düşünelim. 49 kolonun girdi, 1 kolon çıktı olacağını düşünelim. Bu tip bir datasette hangi kolonun feature olarak eklenebileceğini düşünmek EDA yapılsa bile zor olabilir.

-- Lasso regresyonu, çıktı ile alakasız olabilecek kolonların katsayısını yani tetasını sıfırlamaya yarar. 

-- Ridge regresyonundan farkı (ridge regresyonu MSE + lamba(penalty) ekliyordu) ağırlıkların karesi yerine mutlak değerinin kullanılmasıdır. Böylelikle bir hedef değişken çıktı ile yeterince ilişkilendirilmediyse, teta yani katsayı 0'a çekilir.

-- Böylece modelden anlamsız bir değişken çıkartılmış olur, daha sade, daha açıklayıcı ve genellenebilir hale gelir.

-- Ridge regresyonunda olduğu gibi lambda'sı iyi ayarlanmalı, dolayısıyla underfittinge sebep olabilir.

##### Elastic Net (FEATURE SELECTION ve OVERFITTING ENGELLER)

-- Hem L1 hem de L2 yani ridge ve lasso eklenmiş olur. Lambdalar daha kritik, çünkü hem ridge hem de lasso lambdaları var.

-- Lambdaların doğru seçilmesi durumunda hem overfit hem de feature selection konusunda avantaj sağlar.

-- Modellemelerde 3 konuyu da deneyip neyin daha faydalı olduğuna r2 score'una bakarak öğreneceğiz.


Sonuç: Ridge, Lasso ve Elastic Net regresyonları, doğrusal modellerin daha kararlı ve anlamlı olmasını sağlar. Özellikle çok sayıda değişken içeren veri setlerinde Lasso ve Elastic Net kullanılması feature selection konusunda fayda sağlar. Modelin doğruluğunu arttırmak için bu yöntemlerin kullanılması ve doğru şekilde seçilmesi gerekmektedir.

##### Çapraz Doğrulama (Cross Validation) Yöntemleri

-- Train test yapıldıktan sonra train içerisinde validasyon yapılması aslında otomatik oluyor. Ancak bu validasyona karışabiliyoruz.

-- Yani eğitim verisinin içinde olan verileri validation yani doğrulama için kullanılabiliyor. 

-- Daha model eğitilirken eğitim verisinin doğruluğunu ölçmek amacıyla yapılır. 

-- Lambda gibi parametreleri doğru bulabilmek için validasyonları kullanmak gerekir.

Çeşitleri;

    * Leave-One-Out(LOOCV)
    * Leave-P-Out
    * K-Fold Cross Validation
    * Stratified K-Fold Cross Validation
    * Time Series Cross Validation


##### Leave-One-Out(LOOCV)

-- Her deneyden verisetinden bir gözlem çıkartılır, kalan verilerle model eğitilir, çıkarılan gözlem ile test yapılır.

    --Özellikleri
        --Avantajları
            * Eğitim için n-1 gözlem, test için 1 gözlem kullanılır. 
            * n gözlem varsa n defa eğitim yapılır.

        --Dezavantajları
            *Çok büyük veri setlerinde yavaş.
            * Overfittinge sebep olabilir.

##### Leave-P-Out(LPOCV)

-- P adet doğrulama için ayırılır. P'yi biz belirliyoruz, LOOCV'nin sayısını biz verebildiğimiz hali bu. 

    --Özellikleri
        --Avantajları
            * Daha fazla veriyle güvenililirlik artar.

        --Dezavantajları
            *Çok büyük veri setlerinde yavaş.


##### K- Fold Cross Validation

-- Veri setini k eşit parçaya böler. Her seferinde bir parça doğrulama için diğerleri eğitim için kullanılır. Bu işlem k kez tekrarlanır. 

-- Dengeli doğrulama sağlar. 

-- İstenilen parçaya da bölünür. 

-- Veri setini 5'e ayırarak işlem yapar. 100 veri var ise 20 20 20 20 20 olarak bölüyor. İlk 20 train içi validation kalan 80 eğitim gibi gibi.

-- En sık kullanılan cross validation yöntemidir.

    --Özellikleri
        --Avantajları
                * Dengeli doğrulama sağlar. 
                * Hesaplama maliyeti LOOCV'ye göre çok düşüktür. 
                * Genellikle 5 eşit parçaya bölünüyor ama onu biz de belirleyebiliyoruz.
