```python
df["bedrooms_per_room"] = np.log1p(df["bedrooms_per_room"])
df["population_per_household"] = np.log1p(df["population_per_household"])
df["rooms_per_household"] = np.log1p(df["rooms_per_household"])
```

```python
df["ocean_proximity"].unique()
```

```python
df = pd.get_dummies(df, columns=["ocean_proximity"], drop_first=True)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df[["longitude", "latitude", "housing_median_age", "total_rooms", "total_bedrooms", "population", "households", "median_income", "rooms_per_household", "population_per_household", "bedrooms_per_room","ocean_proximity_INLAND", "ocean_proximity_ISLAND", "ocean_proximity_NEAR BAY", "ocean_proximity_NEAR OCEAN"]]
y = df["median_house_value"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state= 42, test_size= 0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
print("mae: ", mae)
print("mse: ", mse)
print("r2: ", r2)
print("rmse: ", np.sqrt(mse))
```
