```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(columns='mpg')
y = df["mpg"]
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, random_state= 42, test_size= 0.2
)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
score = r2_score(y_test, y_pred)
```

```python
print("mae: ", mae)
print("mse: ", mse)
print("r2: ", score)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
# Katsayıları ve özellik isimlerini birleştirme
coefficients = pd.DataFrame(model.coef_, X.columns, columns=['Coefficient'])

# Anlamayı kolaylaştırmak için değerleri büyükten küçüğe sıralayalım
coefficients = coefficients.sort_values(by='Coefficient', ascending=False)

print(coefficients)
```

```python
print("mae: ", mae)
print("mse: ", mse)
print("r2: ", score)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
model_scaled = LinearRegression()
model_scaled.fit(X_train_scaled, y_train)
```

```python
scaled_coefficients = pd.DataFrame(model_scaled.coef_, X.columns, columns=['Scaled Coefficient'])
print(scaled_coefficients.sort_values(by='Scaled Coefficient', ascending=False))
```

```python
y_pred_scaled = model_scaled.predict(X_test_scaled)
```

```python
mae = mean_absolute_error(y_test, y_pred_scaled)
mse = mean_squared_error(y_test, y_pred_scaled)
score = r2_score(y_test, y_pred_scaled)
```

```python
print("mae: ", mae)
print("mse: ", mse)
print("r2: ", score)
```
